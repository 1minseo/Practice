```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.parse import quote #quote 사용하려면 필요
```

```python
baseurl='https://www.siksinhot.com/search'
url=f'{baseurl}?keywords={quote("신촌역")}'
res=requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
lis=soup.select('.localFood_list > li') #여기서부터 다시
print(len(lis)) #리스트 개수
```

```python
soup
```

```python
li=lis[0] #첫번째 리스트를 li변수에 넣어줌
li
```

```python
atag=li.select_one('figcaption>a')
name=atag.select_one('h2').get_text().strip()
name
```

```python
menu=li.select('.cate>a')[-1].get_text().strip() #1도 가능
menu #[1]? '신촌/이대', '김밥' 중 '김밥' 선택
```

```python
score=atag.select_one('.score').get_text().strip()
score
```

```python
sub_href=atag.get('href') #서브페이지 만든 후 mini_soup만들기
#atag['href] 또는 atag.attrs('href)
sub_href
```

```python
#mini_soup
mini_res=requests.get(sub_href)
mini_soup=BeautifulSoup(mini_res.text, 'html.parser') #변수만 mini로 바꿔줌
mini_soup
```

```python
addr=li.select_one('figure>a>img').get('alt').split(',')[-1]
addr
#'전화번호'는 자바스크립트로 사이트가 개편돼서 가져올 수 없음
```

```python
data=[]
for li in lis:
    atag=li.select_one('figcaption>a')
    name=atag.select_one('h2').get_text().strip()
    menu=li.select('.cate>a')[-1].get_text().strip()
    score=atag.select_one('.score').get_text().strip()
    addr=li.select_one('figure>a>img').get('alt').split(',')[-1]
    data.append({'업소명':name, '평점':score, '메뉴':menu, '주소':addr})
data    
```

```python
df=pd.DataFrame(data) #데이터프레임으로 만들기
df.head()
df.to_csv("식신신촌역.csv")
```
