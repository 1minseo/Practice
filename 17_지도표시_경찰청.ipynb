```python
from bs4 import BeautifulSoup
import pandas as pd #데이터 처리를 위한
import time
from selenium import webdriver #웹 스크래핑을 위한
from selenium.webdriver.common.keys import Keys #Keys클래스를 가져와 키보드 키를 조작하는데 사용
from selenium.webdriver.common.by import By #By클래스를 가져와 HTML요소를 찾는데 사용
import folium #지도시각화를 위한
from urllib.parse import quote #URL 인코딩을 위한
import json #JSON데이터 처리를 위한
import requests #HTTP 요청을 위한
```

```python
with open('./kakao.txt') as file: #카카오키 가져오기
    kakao_key=file.read()
```

```python
#기관소개 ->전국경찰관서안내
url='https://www.police.go.kr/user/bbs/BD_selectBbsList.do?q_bbsCode=1038&q_tab=1'
driver = webdriver.Chrome() #크롬드라이버 불러오기
driver.get(url) #크롬드라이버 연결
```

```python
#서울 버튼(id="tab_1") 클릭
driver.find_element(By.ID,'tab_1').click() #find_element: 웹페이지에서 요소(Element)를 찾는데 사용
time.sleep(3) #프로그램을 3초 동안 대기

#By.ID: HTML 요소의 ID 속성으로 요소를 찾음
#By.CLASS_NAME: HTML 요소의 클래스명으로 요소를 찾음
#By.XPATH: XPath를 사용하여 요소를 찾음
```

```python
#soup으로 페이지 정보 가져오기
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup
```

```python
#tbody태그 > tr태그: 리스트를 가져올 수 있음
lis=soup.select('tbody > tr')
len(lis) #리스트 개수
```

```python
li=lis[0] #첫번째 리스트 가져오기
li
#li=lis[-1]: 마지막 리스트 가져오기
```

```python
#관서 가져오기
title=li.select('.show-col')[0].text #이름은 3개 리스트 중 첫번째. 선택된 요소의 텍스트를 추출
title
```

```python
#주소 가져오기
addr=li.select('.show-col')[1].text
addr
```

```python
#홈페이지 가져오기
href=li.select('.show-col')[2].text
href
```

```python
#서울 정보만
lines=[] #각 요소의 정보를 담을 빈 리스트 생성
for li in lis: #반복문. lis에 있는 각 요소를 li라는 변수로 순회
    title=li.select('.show-col')[0].text
    addr=li.select('.show-col')[1].text
    href=li.select('.show-col')[2].text
    lines.append([title,addr,href])
lines    
```

```python
df=pd.DataFrame(lines, columns=['관서명','주소','홈페이지']) #lines리스트의 데이터를 포함하는 데이터프레임 생성. 컬럼명 지정
df.head() #처음 5개 행 추출
```

```python
#지도 불러오기
lat_list, lng_list = [], []
for i in df.index:
    base_url = 'https://dapi.kakao.com/v2/local/search/address.json'
    header={'Authorization': f'KakaoAK {kakao_key}'}
    url = f'{base_url}?query={quote(df["주소"][i])}'
    result = requests.get(url, headers=header).json()
    try:
        lng_list.append(result['documents'][0]['x'])
        lat_list.append(result['documents'][0]['y'])
    except:
        print(df.관서명[i]) #이 부분만 변경하면 됨. 나머지는 복붙

#DataFrame(df)에서 주소를 추출하여 카카오맵 API를 사용해 해당 주소의 위도(lat), 경도(lng) 가져오기
```

전국 경찰서 정보 가져오기

```python
#전국 지역의 수 18개. 서울 -> 전국 순으로
count=len(soup.select('.nav.nav-tabs.ui-multi-line > li')) #박스를 감싸고 있는 태그 찾기. 클래스명 공백에 '.'찍기
count
```

```python
#전체 for문 돌리기 (18번)
lines=[]
for i in range(1, count + 1): #n-1. 1부터 count까지의 숫자 범위(=웹페이지의 탭의 수)를 반복
    driver.find_element(By.ID, f'tab_{i}').click()
    time.sleep(3)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    lis=soup.select('tbody > tr')
    for li in lis: #i가 아니라 li. 밑에랑 일치해야함
        title=li.select('.show-col')[0].text
        addr=li.select('.show-col')[1].text
        href=li.select('.show-col')[2].text
        lines.append([title,addr,href])
lines       
```

```python
#lines 리스트의 데이터를 이용해 새로운 DataFrame 생성
df=pd.DataFrame(lines,columns=['관서명','주소','홈페이지']) #관서명,주소,홈페이지 정보를 담고 있는 DataFrame 생성
df.head()
```
