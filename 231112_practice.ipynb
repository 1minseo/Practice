```python
df['species_3'] = df['species'].apply(lambda x : x[:3]) #apply(): 데이터를 변환하거나 사용자가 정의한 함수를 적용
df
# df['species']: 데이터프레임에서 'species' 열 선택
# apply(lambda x: x[:3]): 'species' 열의 각 값에 대해 첫 세글자를 추출하는 람다함수(익명함수) 적용
# df['species_3']: 새로운 'species_3' 열 생성 후 추출된 세글자로 채워넣음
```

```python
def sample(x):
    x = x[-3:] #뒤에서 3번째까지의 문자를 가져오는 함수
    return x
```

```python
df['species-3'] = df['species'].apply(sample)
df
```
