```python
#구조 외우기#
import requests
from bs4 import BeautifulSoup
url= 'https://news.naver.com/main/ranking/popularDay.naver'
headers={"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'}
res=requests.get(url, headers=headers)
soup=BeautifulSoup(res.text,'html.parser')
import re
```

```python
soup
```

```python
newslist=soup.select(".rankingnews_list")
len(newslist)
```

```python
items=soup.select('ul.rankingnews_list>li')
items
```

```python
news=newslist[0]
news
```

```python
lis=news.find_all('li')
lis[0]
```

```python
news_rank=lis[0].select_one(".list_ranking_num").text
news_rank
```

```python
news_title=lis[0].select_one(".list_title") #title=제목+링크
news_title_text=news_title.text
```

```python
news_link=news_title.get('href') #link=링크만
news_link
```

```python
newsData=[] #빈 리스트(newsData)를 생성
for news in newslist[:16]: #newslist 안의 뉴스 정보를 하나씩 확인
    lis=news.find_all('li') #각 뉴스 안에서 'li'태그 모두 찾기
    print(lis)
    for li in lis: #각 'li'태그를 하나씩 확인
        news_rank=li.select_one(".list_ranking_num").text #해당 항목의 순위 가져오기
        news_title=lis[0].select_one(".list_title") #해당 항목의 제목 요소 찾기
        news_title_text=news_title.text #제목 텍스트를 추출하여 변수에 할당
        news_link=news_title.get('href')
        newsData.append({"ranking":news_rank,
                         "title":news_title_text,
                         "link":news_link})
```

```python
newsData
```

```python
len(newsData)
```

```python
#뉴스 작성 시간
news_url=newsData[0]['link']
news_url
```

```python
mini_res=requests.get(news_url, headers=headers)
mini_soup=BeautifulSoup(mini_res.text,'html.parser') #메인페이지X. 미니페이지
news_time=mini_soup.select_one(".media_end_head_info_datestamp").select_one(".media_end_head_info_datestamp_time").get('data-date-time')
print(news_time)
```

```python
news_contents=mini_soup.select_one("#newsct_article").text.replace("\n","").replace("\t","") #본문
news_contents
```

```python
for news in newsData:
    news_url = news['link']
    mini_res = requests.get(news_url, headers=headers)
    mini_soup= BeautifulSoup(mini_res.text, 'html.parser')
    news_time = mini_soup.select_one(".media_end_head_info_datestamp").select_one(".media_end_head_info_datestamp_time").attrs['data-date-time']
    news_contents = mini_soup.select_one("#newsct_article").text.replace("\n","").replace("\t","")
    news['time'] = news_time
    news['contents'] = news_contents
```
