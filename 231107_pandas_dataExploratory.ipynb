```python
import pandas as pd #판다스 라이브러리 import
df = pd.read_csv("/content/auto-mpg.csv", header=None)

#pd.read_csv(파일명 or 경로)
#header(열이름)이 없을 경우 header=None, 1번째 행이 열이름인 경우 header=1
```

```python
# 열이름 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name']
```

```python
# df 의 모양 크기를 확인
df.shape
```

```python
# df의 내용확인
df.describe(include="all") # 평균,  표준편차, 최대최소, 4분위수 기술통계

#df.describe(): 총 데이터 수(count), 평균(mean), 표준편차(std), 분위수(25,50%(=중앙값),75%), 최대최소(max,min)
```

```python
df3 = df2.iloc[[0, 5], 2:].head()

# iloc를 사용해 특정 행과 열 선택 -> [행([0, 5]), 열(2:)] 2개의 리스를 사용해 행과 열 선택
```

```python
df3
```

```python
df3.index = ['south', 'north'] # 행인덱스 변경
```

```python
df3.columns = df3.columns.map(int) #map(int): 판다스 시리즈 객체의 요소를 정수로 변환하는 메서드
```

```python
df3.info()
```

```python
# 선그래프 그리기
df3.plot()
```

```python
# 행, 열 전치하여 다시그리기
tdf3 = df3.T
print(tdf3)
```

```python
tdf3.plot()
```

```python
df3 = df3.astype('float') #astype:df 또는 시리즈의 데이터 형식을 변경
tdf3 = df3.T
tdf3.head()
tdf3.plot(kind='bar')

#df3 = df3.astype('float'): df3의 모든 열을 float형식으로 변환
#tdf3 = df3.T: df3 전치 -> tdf3 생성
#tdf3.head(): tdf3의 처음 몇 개의 행을 출력
```
